```python
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import warnings

warnings.filterwarnings("ignore")
    
folder = 'D:\\Internships\\aigora\\jupyter notebooks\\audio_chunks'
file = 'test.wav'
```

```python
samples, sample_rate = librosa.load(folder+'//'+file, sr=16000)
time = np.arange(0, len(samples))/sample_rate
```

```python
fig, ax = plt.subplots()
ax.plot(time, samples)
ax.set(xlabel='time', ylabel='amplitude')
plt.show()
```
